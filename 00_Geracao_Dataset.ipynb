```
from datasets import load_dataset

dataset = load_dataset("SetFit/amazon_massive_scenario_pt-PT")
```

```
import pandas as pd

df = pd.DataFrame(dataset['train'])
df.head()
```

```
df.columns
```

```
df_documentos = df[['text']].copy()

df_documentos.columns = ['documento']
df_documentos.insert(0, 'id', range(1, len(df_documentos) + 1))

df_documentos = df_documentos.head(50)

df_documentos.head()
```
